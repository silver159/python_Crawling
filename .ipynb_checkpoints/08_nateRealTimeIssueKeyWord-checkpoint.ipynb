```python
import warnings
warnings.filterwarnings(action='ignore')
%config Completer.use_jedi = False 
import requests
from bs4 import BeautifulSoup
```

```python
targetSite='https://www.nate.com/'
hearders={'User-agent': 'Mozilla/5.0 (Windows NT 6.3; WOW64; Trident/7.0; rv:11.0) like Gecko'}
request = requests.get(targetSite, headers=hearders)
# print(request)
html = request.text
soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

```python
# 순위 크롤링
# ranks = soup.findAll('span',{'class':'num_rank'})
# print(ranks[:5])
# for rank in ranks[:5]:
    # print(rank.text)

# ranks = soup.select('#olLiveIssueKeyword .num_rank')
# 오른쪽 클릭 => copy => selector
ranks = soup.select('#olLiveIssueKeyword > li > span')
# print(ranks)
# for rank in ranks:
    # print(rank.text)
 
# 실시간 이슈 키워드와 상승/하락 폭
issues = soup.findAll('a',{'class':'ik'})
# print(issues)
# for issue in issues:
    # print(issue.text.strip().split('\n'))
    # print(issue.text.strip().split('\n')[0], issue.text.strip().split('\n')[1])

for i in range(5):
    rank = ranks[i].text
    issue = issues[i].text.strip().split('\n')[0]
    upDown = issues[i].text.strip().split('\n')[1]
    # print('%2s위: %s' % (rank, issue))
    print('{:>2}위: {}'.format(rank, issue),end='')
    if upDown[:2] == '상승':
        print('[{}{}]'.format('▲',upDown[-1]))
    elif upDown[:2] == '하락':
        print('[{}{}]'.format('▼',upDown[-1]))
    elif upDown[:2] == '동일':
        print('[-]')
    else:
        print('[new]')
```
